### ADR Plot- City Hotel Vs Resort Hotel

##### To find the price fluctuation relationship between City and Resort Hotels.

```python
# Grouping by date
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20, 8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize=30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel')
plt.legend()
plt.show()
```

##### 1. Here we can find that Resort hotel prices fluctuate more than City Hotels, which has a lower fluctuation.
##### 2. Both prices fluctuate but Resort Hotel prices go up much more than city hotels from June to September.
##### 3. An Increase in price can be seen in resort hotels during January most probably due to New Year.
##### 4. The spikes we see in the above diagram states the prices of are high.

### Multi line Chart 

#### Daily Average Rate vs Cancelled and Non Cancelled bookings

```python
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], dayfirst=True)
cancelled_data = df[df['is_canceled'] == 1]
not_cancelled_data = df[df['is_canceled'] == 0]

cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)
```

```python
cancelled_df_adr=cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016') & ( cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr=not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016') & (not_cancelled_df_adr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate', fontsize=30)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='cancelled')
plt.legend(fontsize=20)
plt.show()
```

##### 1. Here we can see that The average daily rate for cancelled is more than not cancelled except seasonal times(June, July)
##### 2. Which means that our hypothesis is right and high price is a factor for high cancellation rates

### Correlation Heatmap

#### To find if there is a positive relation between waiting period and cancellation

```python
# Compute correlation matrix for relevant numerical columns
correlation_matrix = df[['days_in_waiting_list', 'is_canceled', 'lead_time', 'adr']].corr()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5, fmt=".2f")
plt.title("Correlation Heatmap of Key Variables")
plt.show()
```

##### 1.The correlation is very weak (close to 0), meaning there is no strong direct relationship between waiting list days and cancellations.
##### 2.This suggests that while longer waiting lists might contribute to cancellations, other factors likely have a stronger influence.
